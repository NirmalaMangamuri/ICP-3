```python
import numpy as np
import matplotlib.pyplot as plt

# Create an empty subplot with 2 rows and 2 columns (4 subplots total)
# Create the same plot as above with 2 rows and 2 columns and figsize of (10, 5)
fig, axs = plt.subplots(2, 2, figsize=(10, 5))

# Prepare data for the plots
X = np.linspace(0, 10, 100)
favorite_foods = {"Pizza": 12.99, "Sushi": 15.50, "Ice Cream": 5.25}

# Plot X versus X/2 on the top left axes
axs[0, 0].plot(X, X/2)
axs[0, 0].set_title("X vs X/2")
axs[0, 0].set_xlabel("X")
axs[0, 0].set_ylabel("X/2")

# Plot a scatter plot of 10 random numbers on each axis on the top right subplot
x_scatter = np.random.rand(10)
y_scatter = np.random.rand(10)
axs[0, 1].scatter(x_scatter, y_scatter)
axs[0, 1].set_title("Scatter Plot of 10 Random Points")
axs[0, 1].set_xlabel("X")
axs[0, 1].set_ylabel("Y")

# Plot a bar graph of the favourite food keys and values on the bottom left subplot
axs[1, 0].bar(favorite_foods.keys(), favorite_foods.values())
axs[1, 0].set_title("Favourite Food Prices")
axs[1, 0].set_xlabel("Food")
axs[1, 0].set_ylabel("Price ($)")

# Plot a histogram of 1000 random normally distributed numbers on the bottom right subplot
data = np.random.randn(1000)
axs[1, 1].hist(data, bins=30, edgecolor='black')
axs[1, 1].set_title("Histogram of Normally Distributed Data")
axs[1, 1].set_xlabel("Value")
axs[1, 1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the uploaded CSV file
car_sales = pd.read_csv('car-sales (1).csv')

# Convert 'Price' column to numeric, coercing errors to NaN
car_sales['Price'] = pd.to_numeric(car_sales['Price'], errors='coerce')

# Add a column with cumulative sum of 'Price'
car_sales['Total Sales'] = car_sales['Price'].cumsum()

# Add a column with a date range starting from today
car_sales['Date Sold'] = pd.date_range(start=pd.Timestamp.today(), periods=len(car_sales), freq='D')

# View the updated DataFrame
print(car_sales.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
# This step is now redundant as Price is already cleaned in a previous cell.
# car_sales['Price'] = car_sales['Price'].str.replace('[^\d.]', '', regex=True).astype(float)

# Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column
# Ensure the 'Price' column is numeric before performing cumsum
if car_sales['Price'].dtype != 'float64':
     print("Warning: 'Price' column is not float64. Attempting to convert.")
     try:
         car_sales['Price'] = car_sales['Price'].str.replace('[^\d.]', '', regex=True).astype(float)
     except AttributeError:
          print("Could not convert 'Price' to float. It might already be numeric.")
     except Exception as e:
          print(f"An error occurred during price conversion: {e}")


car_sales['Total Sales'] = car_sales['Price'].cumsum()

# Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
# Ensure the length matches the DataFrame length
car_sales['Sale Date'] = pd.date_range(start=pd.Timestamp.today(), periods=len(car_sales), freq='D')

# View the car_sales DataFrame
print(car_sales.head())

# Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
car_sales.plot(x='Sale Date', y='Total Sales')

plt.title("Total Sales Over Time")
plt.xlabel("Sale Date")
plt.ylabel("Total Sales")
plt.show()

# Convert the 'Price' column to the integers
# Ensure 'Price' is numeric and handle NaN values before converting to int
if car_sales['Price'].dtype == 'float64':
    # Fill NaN values with 0 before converting to int
    car_sales['Price'] = car_sales['Price'].fillna(0).astype(int)
else:
    print("Warning: 'Price' column is not float64, cannot convert to int.")


# Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function
car_sales.plot.scatter(x='Odometer (KM)', y='Price')

plt.title("Scatter Plot of Odometer vs Price")
plt.xlabel("Odometer (KM)")
plt.ylabel("Price")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Import "../data/heart-disease.csv" and save it to the variable "heart_disease"
heart_disease = pd.read_csv('heart-disease (1).csv')

# View the first 10 rows of the heart_disease DataFrame
print(heart_disease.head(10))

# Create a histogram of the "age" column with 50 bins
heart_disease['age'].hist(bins=50)
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Call plot.hist() on the heart_disease DataFrame and toggle the "subplots" parameter to True
heart_disease.plot.hist(subplots=True, figsize=(10, 12))
plt.show()
```

```python
# Call plot.hist() on the heart_disease DataFrame with subplots=True and figsize=(10, 30)
heart_disease.plot.hist(subplots=True, figsize=(10, 30))
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame with patients over 50 years old
over_50 = heart_disease[heart_disease['age'] > 50]

# Create the plot
fig, axes = plt.subplots(nrows=len(over_50.columns), figsize=(10, 30))

# Plot the data
for i, column in enumerate(over_50.columns):
    if over_50[column].dtype in ['int64', 'float64']:  # plot only numeric columns
        over_50[column].plot.hist(ax=axes[i], bins=50, alpha=0.7)
        # Add mean line
        mean_value = over_50[column].mean()
        axes[i].axvline(mean_value, color='red', linestyle='dashed', linewidth=2)
        axes[i].set_title(f'Histogram of {column}')
        axes[i].set_xlabel(column)
        axes[i].set_ylabel('Frequency')
    else:
        axes[i].set_visible(False)  # hide non-numeric plots

plt.tight_layout()
plt.show()
```
